```python
import pandas as pd
import reverse_geocode
```

```python
df = pd.read_csv("significant_earthquakes1965-2016.csv",sep=",")
df.head()
```

```python
def obtain_country(lat,long):
    c = reverse_geocode.search([(lat,long)])[0]["country"]
    return c
```

```python
df["Country"] = df.apply(lambda x: obtain_country(x["Latitude"],x["Longitude"]),axis=1)
df.head()
```

```python
pd.set_option('display.max_rows', 500)
df["Country"].value_counts()
```

```python
# export
df.to_csv("significant_earthquakes1965-2016_with_countries.csv",sep=",")
```
